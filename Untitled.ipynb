```python
import pandas as pd
import numpy as np

patient_data = pd.read_csv("data/patient_data.csv")
```

```python
days_of_week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]

def score_bound(x):
    if x > 1:
        return 1
    elif x < 0:
        return 0
    else:
        return x

for i, row in patient_data.iterrows():
    patient_data.at[i, "Day of Week"] = days_of_week[np.random.randint(0,4)]
    patient_data.at[i, "Days in Advance"] = np.random.randint(1, 4)
    
    if patient_data.at[i, "No Show"] == 0:
        score = np.random.normal(0.1, 0.2)
        patient_data.at[i, "Previous No-Show Coefficient"] = score_bound(score)
    else:
        score = np.random.normal(0.2, 0.2)
        patient_data.at[i, "Previous No-Show Coefficient"] = score_bound(score)
```

```python
patient_data.insert(2,"Sex", 1, True)
patient_data.insert(3, "Day", 1, True)
#patient_data = patient_data.drop(patient_data.columns[1], axis = 1)

for i, row in patient_data.iterrows():
    patient_data.at[i, "Day"] = days_of_week.index(patient_data.at[i, "Day of Week"])
    if patient_data.at[i, "Gender"] == 'F':
        patient_data.at[i, "Sex"] = 0
    else:
        patient_data.at[i, "Sex"] = 1
patient_data = patient_data.drop(patient_data.columns[1],axis = 1)
patient_data = patient_data.drop(patient_data.columns[3],axis = 1)

patient_data.rename({"Sex": "Gender", "Day": "Day of Week"})
patient_data
```
